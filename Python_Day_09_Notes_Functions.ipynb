```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ajaverett/data/refs/heads/main/titanic.csv')
```

```python
df
```

```python
def titanic_age_now(age):
  titanic_year = 1912
  current_year = 2025
  birth_year = titanic_year - age
  how_old_the_would_be_now = current_year - birth_year
  return how_old_the_would_be_now
```

```python
titanic_age_now(20)
```

```python
titanic_age_now(42)
```

```python
df["Age_now"] = df["Age"].apply(titanic_age_now)
```
